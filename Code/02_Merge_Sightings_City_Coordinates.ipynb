```python
import pandas as pd
from pathlib import Path
import numpy as np
import time
import requests
```

```python
sightings_file = pd.read_csv('../Output/usa_sightings_df_.csv')
complete_file = pd.read_csv('../Output//city Data Complete.csv')
```

```python
sightings_df = pd.DataFrame(sightings_file)
complete_df = pd.DataFrame(complete_file)
```

```python
complete_df1=complete_df.drop(['Unnamed: 0'], axis = 1)
```

```python
complete_df1  =complete_df1.rename(columns={'City': 'CITY'})
complete_df1
```

```python
#performing right join.
city_complete_df = pd.merge(sightings_df,  
                      complete_df1,  
                      on ='CITY', 
                      how ='right') 
```

```python
city_complete_df.head(20)
```

```python
# Convert SIGHTING_ID and YEAR to integers
city_complete_df['SIGHTING_ID'] = city_complete_df['SIGHTING_ID'].fillna(0).astype(int)
city_complete_df['YEAR'] = city_complete_df['YEAR'].fillna(0).astype(int)
```

```python
# Remove NaN values from Lat and Lng columns
city_complete_df['Lat'] = city_complete_df['Lat'].fillna(0.0) 
city_complete_df['Lng'] = city_complete_df['Lng'].fillna(0.0)
```

```python
# Drop rows where both Lat and Lng are NaN
city_complete_df = city_complete_df.dropna(subset=['Lat', 'Lng'])
city_complete_df = city_complete_df.dropna(subset=['CITYSTATE'])
```

```python
# Round Lat / Lng for processing performance improvement in mapping
city_complete_df['Lat'] = city_complete_df['Lat'].round(2)
city_complete_df['Lng'] = city_complete_df['Lng'].round(2)
```

```python
# Sort dataframe
city_complete_df = city_complete_df.sort_values(by=["OCCURRED_DATE"], ascending=False)
```

```python
# Removing duplicates
city_complete_df = city_complete_df.drop_duplicates()
```

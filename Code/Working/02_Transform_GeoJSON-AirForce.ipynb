```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import json
```

```python
# Load data
df = pd.read_json('Output/all_us_air_force_bases.json')  # Ensure your CSV has 'latitude' and 'longitude' columns
df
```

```python
# Create GeoDataFrame
gdf = gpd.GeoDataFrame(
    df,
    geometry=[Point(xy) for xy in zip(df['longitude'], df['latitude'])],
    crs='EPSG:4326'  # EPSG:4326 is the coordinate reference system for WGS84
)
gdf
```

```python
# Output to GeoJSON
geojson = gdf.to_json()
```

```python
# Pretty-print the GeoJSON
geojson_pretty = json.dumps(json.loads(geojson), indent=4)
```

```python
# Save to file
with open('airforce_pretty.geojson', 'w') as f:
    f.write(geojson_pretty)

print("GeoJSON data has been pretty-printed and saved to 'locations_pretty.geojson'")
```

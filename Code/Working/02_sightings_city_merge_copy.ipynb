```python
import pandas as pd
from pathlib import Path
```

```python
sightings_file = Path("Output/usa_sightings_df_.csv")
complete_file = Path("Output/city Data Complete.csv")
```

```python
sightings_df = pd.read_csv(sightings_file)
complete_df = pd.read_csv(complete_file)
```

```python
sightings_df
```

```python
complete_df1 = complete_df.drop(['Unnamed: 0'], axis = 1)
```

```python
complete_df1  =complete_df1.rename(columns={'City': 'CITY'})
complete_df1
```

```python
#performing right join.
city_complete_df = pd.merge(sightings_df,  
                      complete_df1,  
                      on ='CITY', 
                      how ='right') 
```

```python
city_complete_df
```

```python
city_complete_df = city_complete_df.dropna()
```

```python
city_complete_df
```

```python
city_complete_df.to_csv('City_latlng_merge.csv')
```
